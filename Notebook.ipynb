### Filter Data
- After read and analyze the data, there is data containing non-countries in the country name section

```python
def filter_real_countries(total_population):
    # Get Codes with 3 Code Country from Pycountry libraries
    valid_country_codes = set(country.alpha_3 for country in pycountry.countries)
    
    # Filter Dataframe by 3 Code Country 
    clean_df = total_population[total_population['Country Code'].isin(valid_country_codes)].copy()
    
    return clean_df
```

```python
# Apply to your dataset
cleaned_df = filter_real_countries(total_population)

print(cleaned_df['Country Name'].unique())
print("\n Total no of unique countries:",cleaned_df['Country Name'].nunique())
```

## Visualization Total Population 2024

```python
# Filter Dataframe
male_df = cleaned_df[cleaned_df['Indicator Code'] == 'SP.POP.TOTL.MA.IN'][['Country Name', '2024']]
female_df = cleaned_df[cleaned_df['Indicator Code'] == 'SP.POP.TOTL.FE.IN'][['Country Name', '2024']]

# Rename Column 2024 -> Male / Female 2024
male_df.rename(columns={'2024' : 'Male 2024'}, inplace=True)
female_df.rename(columns={'2024' : 'Female 2024'}, inplace=True)

# Merge the Male and Female DataFrame by Country Name
total_2024 = pd.merge(male_df,female_df, on='Country Name')

# Sum the Male and Female and create a New Column "Total Population"
total_2024['Total Population 2024'] = total_2024['Male 2024'] + total_2024['Female 2024']

# Sort it by Total Population
top10 = total_2024.sort_values('Total Population 2024', ascending=False).head(10)

# This Code change the value to string 
top10[['Male 2024','Female 2024', 'Total Population 2024']] = top10[['Male 2024','Female 2024', 'Total Population 2024']].applymap(lambda x: f"{x:,.0f}")

top10
```

```python
# Convert back to numbers for visualization
top10_plot = total_2024.sort_values('Total Population 2024', ascending=False).head(10)

# Set style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(15,6))
sns.barplot(data=top10_plot, x= "Total Population 2024", y= "Country Name", palette='viridis', legend=False, hue="Country Name")

# Labels and title
plt.title('Top 10 Countries by Total Population in 2024', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Total Population (2024)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Top 5 Growth population for 5 years

```python
# Filter DataFrame
male_df = cleaned_df[cleaned_df['Indicator Code'] == 'SP.POP.TOTL.MA.IN'][['Country Name', '2020','2021','2022','2023','2024']]
female_df = cleaned_df[cleaned_df['Indicator Code'] == 'SP.POP.TOTL.FE.IN'][['Country Name', '2020','2021','2022','2023','2024']]

# Rename Columns
male_df.columns = ['Country Name', 'Male 2020', 'Male 2021', 'Male 2022', 'Male 2023', 'Male 2024']
female_df.columns = ['Country Name', 'Female 2020', 'Female 2021', 'Female 2022', 'Female 2023', 'Female 2024']

# Merge DataFrame
merged_df = pd.merge(male_df, female_df, on='Country Name')

# Calculate total population for each year
for year in range(2020, 2025):
    merged_df[f'Total {year}'] = merged_df[f'Male {year}'] + merged_df[f'Female {year}']

# Create list with year
year_cols = [f'Total {y}' for y in range(2020, 2025)] #Output -> Total 2020, Total 2021, etc

# Country name columns + year_cols assign as a new value 
growth_data = merged_df[['Country Name'] + year_cols].copy()

# Compute year-by-year growth
growth_data['2021'] = growth_data['Total 2021'] - growth_data['Total 2020']
growth_data['2022'] = growth_data['Total 2022'] - growth_data['Total 2021']
growth_data['2023'] = growth_data['Total 2023'] - growth_data['Total 2022']
growth_data['2024'] = growth_data['Total 2024'] - growth_data['Total 2023']

# Assign only year-over-year columns
growth_data = growth_data[['Country Name', '2021', '2022', '2023', '2024']]

# Calculate average yearly growth
growth_data['Average Growth'] = growth_data[['2021', '2022', '2023', '2024']].mean(axis=1)

# Get Top 10 countries by average growth
top10_growth  = growth_data.sort_values(by='Average Growth', ascending=False).head(10)

top10_growth[['2021','2022','2023','2024']] = top10_growth[['2021','2022','2023','2024']].applymap(lambda x: f"{x:,.0f}")

print("Top 10 countries by average yearly growth:")
top10_growth
```

```python
# Make sure top10_growth is still a DataFrame
top10_countries = top10_growth['Country Name'].tolist()

# Create a copy window dataset from this subset, so it can modify what we needs and no warning because it's our new real dataframe
plot_data = growth_data[growth_data['Country Name'].isin(top10_countries)].copy()

# Melt to long format
melted_plot = plot_data[['Country Name', '2021', '2022', '2023', '2024']].melt(
    id_vars='Country Name', var_name='Year', value_name='Yearly Growth'
)
melted_plot['Year'] = melted_plot['Year'].astype(int)

# Why using melt ? Because the Line chart need a Long format datasets


plt.figure(figsize=(14, 7))
sns.lineplot(data=melted_plot, x='Year', y='Yearly Growth', hue='Country Name', marker='o')

# Loop each unique country name to draw 1 line per country
# for country in melted_plot['Country Name'].unique():
#     # Filter dataframe to get only the rows for that country, so we can plot its growth trend over time 
#     country_data = melted_plot[melted_plot['Country Name'] == country]
#     plt.plot(country_data['Year'], country_data['Yearly Growth'], marker='o', label=country, linestyle='-')

plt.title('Yearly Population Growth Trend (Top 10 Avg Growth, 2021â€“2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Population Growth')
plt.xticks([2021, 2022, 2023, 2024])
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
